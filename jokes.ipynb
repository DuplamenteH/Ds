```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



%matplotlib inline
sns.set_style('darkgrid')
```

```
url='https://raw.githubusercontent.com/DuplamenteH/Ds/master/high_diamond_ranked_10min.csv'
data = pd.read_csv(url)
data.head()
```

```
cols = ['redTotalExperience', 'redTotalJungleMinionsKilled',
        'redTotalMinionsKilled','blueGoldPerMin','redGoldPerMin',
        'redCSPerMin','blueTotalJungleMinionsKilled','blueTotalMinionsKilled',
        'blueTotalExperience','gameId','blueCSPerMin']
data_modf = data.drop(cols, axis=1)
data_modf
```

```
firstbloodsBlue=data_modf['blueFirstBlood'].shape[0]

blue = 0
for i in range(firstbloodsBlue):
    if(data_modf['blueFirstBlood'][i]==1):
        blue=blue+1
print(blue)

firstbloodsRed=data_modf['redFirstBlood'].shape[0]

red = 0
for i in range(firstbloodsRed):
    if(data_modf['redFirstBlood'][i]==1):
        red=red+1
print(red)


X = [blue,red]
```

```

labels =  'BLUETEAM','REDTEAM'


fig1, ax1 = plt.subplots(figsize=(10,10))
ax1.pie(X, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
ax1.legend()

plt.show()

```

```
blueTorresDestruidas = data_modf['blueTowersDestroyed'].sum()
redTorresDestruidas = data_modf['redTowersDestroyed'].sum()

print(blueTorresDestruidas)
print(redTorresDestruidas)
```

```
t=[blueTorresDestruidas,redTorresDestruidas]


labels =  'BLUETEAM','REDTEAM'


fig1, ax1 = plt.subplots(figsize=(10,10))
ax1.pie(t, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
ax1.legend()

plt.show()
```
